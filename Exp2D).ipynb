```
Name=[]
Marks=[]
Date=[]

def get_name():
  global Name
  global Marks
  global Date
  n = int(input("Enter number of students : ")) 
  for i in range(0, n):

    ele = input(f'Enter Student {i+1} Name') 
  
    Name.append(ele) 
  Marks=Name.copy()
  Date=Name.copy()
  for i in range(0, n):
    Marks[i]=eval(input(f"Enter marks for {Name[i]} (Physics,chemistry,Biology,maths) "))

  for i in range(0, n):
    Date[i]=input(f"Enter students’ admission date in the form (dd/mm/yyyy) {Name[i]} ")



get_name()

print(Marks)

def cal(a):
  c=Name.index(a)
  b=Marks[c]
  b=list(b)
  print(f'the sum of marks are {sum(b)}')
  print(f'the avg of marks are {sum(b)/4}')

#cal('krishna')

def details():
  print('The details of students are as follow')
  for i in range(0,len(Name)):
    print(f'Student No.{i+1} Name:{Name[i]}\n Marks of (Physics,chemistry,Biology,maths):{Marks[i]}\n Date of joining {Date[i]}')

def year():
  for i in range(0,len(Date)):
    for j in range(i+1,len(Date)):
      if Date[i][4:]==Date[j][4:]:
        print(f'{Name[i]}{Date[i]}')
        print(f'{Name[j]}{Date[j]}')
      
def getc():
 print("Menu***************\n1.Add students’ marks information and Date of joining \n2.Display students Details\n3.Display students having admission in the same year\n4. Calculate the total and average marks.of a student \n5.EXIT")
 b=int(input("Enter your choice"))
 return b
c=getc()


while (c>0 and c<=4):
 if c==1:
   get_name()
   c=getc()
    

 elif c==2:
   details()
   c=getc()

 

 elif c==3:

   year()
   c=getc()


 elif c==4:
   print(f'The List of students are{Name}')
   a=input("Enter Name of student to  get the sum and avg of his marks  ")
   cal(a)
   c=getc()
      

```
