**F. Write a menu-driven program to demonstrate the use of dictionary in
python:**

```
dic={}
while True:
    print("Menu$$$$$$$$$$")
    print("1.Enter Players")
    print("2.Display Player Details")
    print("3.Add Player")
    print("4.Remove a player details")
    print("5.Display runs of a player")
    print("6.EXIT")
    choice=int(input("Enter your choice:"))
    if choice ==1:
      global dic
      n=int(input("Enter the number of players"))
      for i in range(n):
        a=input(f"Name of player{i+1}")
        b=input(f"Enter the runs of player{i+1}")
        dic.update({a:b})
      
      print(dic)
      
        
    elif choice==2:

      print(f"The team details are \n{dic}")
      
    elif choice==3:

      a=input(f"Name of player")
      b=input(f"\nEnter the runs of player")
      dic.update({a:b})
      print(dic)

      

    elif choice==4:
      a=input(f"enter the player name from this list \n{dic.keys()}")
      dic.pop(a)
      print(dic)

    elif choice==5:
      a=input(f"enter the player name from this list\n{dic.keys()}")
      
      print(dic[a])
      

    elif choice==6:
      break

    else:
        print('wrong choice')
```
